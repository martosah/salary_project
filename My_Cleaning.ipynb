## 1. Get the Average Salary

```python
data.head()
```

```python
data["Salary Estimate"]=data["Salary Estimate"].apply(lambda x: x.split("(")[0])
```

```python
data["Salary Estimate"]=data["Salary Estimate"].apply(lambda x: x.replace("K","").replace("Employer Provided Salary:","").replace("Per Hour","").replace("$",""))
```

```python
data.tail(50)
```

```python
data=data.loc[data["Salary Estimate"]!="-1"]
```

```python
data["minsal"]=data["Salary Estimate"].apply(lambda x: int(x.split("-")[0]))

data["maxsal"]=data["Salary Estimate"].apply(lambda x: int(x.split("-")[1]))
```

```python
data["Average_salary"]=(data["maxsal"]+data["minsal"])/2
```

## 2. Get the Company name

```python
data["Company Name"]=data["Company Name"].apply(lambda x: x if "\n" not in x else x.split("\n")[0])
```

```python
data.head()
```

## 3. State

```python
data["Location"]=data["Location"].apply(lambda x: x.split(",")[1])
```

```python
data["Location"].value_counts()
```

```python

data["Location"]=data["Location"].apply(lambda x: "CA" if x.strip().lower()=="los angeles" else x.strip())
```

## 4. Job in HQ?

```python
data["Job in HQ"]=data.apply(lambda x: 1 if x["Location"] in x["Headquarters"] else 0, axis=1)
```

#### We want to group the Job title into
1. Data Scientist
2. Data Engineer
3. Data Analyst
4. MLE
5. Manager
6. Director

```python
data["Jobtitle_cat"]=data["Job Title"].apply(lambda x: "data scientist" if "data scientist" in x.lower() else "data engineer" if "data engineer" in x.lower() else "data analyst" if "data analyst" in x.lower() else "mle" if "machine learning" in x.lower() else "manager" if "manager" in x.lower() else "director" if "director" in x.lower() else "na")
```

```python
data["Jobtitle_cat"].value_counts()
```

```python
def jobtitle_grp(x):
    if "data scientist" in x.lower():
        return "data scientist"
    if "data engineer" in x.lower():
        return "data engineer"
    if "analyst" in x.lower():
        return "data analyst"
    if "machine learning" in x.lower():
        return "machine learning"
    if "manager" in x.lower():
        return "manager"
    if "director" in x.lower():
        return "director"
    else:
        return "na"
```

```python
data["Jobtitle_grp"]=data["Job Title"].apply(jobtitle_grp)
```

```python
data["Jobtitle_grp"].value_counts()
```

```python
def sen(x):
    if "sr" in x.lower() or "sr." in x.lower() or "snr" in x.lower() or "snr." in x.lower() or "principal" in x.lower() or "lead" in x.lower():
        return "senior"
    elif "jr" in x.lower() or "jr." in x.lower() or "jnr" in x.lower() or "jnr." in x.lower():
        return "junior"
    else:
        return "na"
```

```python
data["Experience_level"]=data["Job Title"].apply(sen)
```

```python
data["Experience_level"].value_counts()
```
